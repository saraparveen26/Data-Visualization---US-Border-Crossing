```python
# Import dependencies
import pandas as pd

# Store file path
csv_path="Resources/Border_Crossing_Entry_Data.csv"

# Read the csv file
data_df = pd.read_csv(csv_path)
data_df
```

```python
# List the colums in the dataset
data_df.columns
```

```python
# Reduce the dataset to the columns needed
reduced_df=data_df[['State', 'Border', 'Date', 'Measure', 'Value']]
reduced_df
```

```python
# Split the Date column
reduced1_df = reduced_df.copy()
reduced1_df[['Jour', 'Time', 'AM/PM']] = reduced1_df['Date'].str.split(" ", expand = True)
reduced1_df
```

```python
# Reduce the dataset to the columns needed
reduced2_df=reduced1_df[['State', 'Border', 'Measure', 'Value', 'Jour']]
reduced2_df
```

```python
# Split the Jour column
reduced3_df = reduced2_df.copy()
reduced3_df[['Month', 'Day', 'Year']] = reduced3_df['Jour'].str.split("/", expand = True)
reduced3_df
```

```python
# Reduce the dataset to the columns needed
reduced3_df=reduced3_df[['State', 'Border', 'Measure', 'Value', 'Month', 'Year']]
reduced3_df
```

```python

# reduced4_df = reduced3_df.loc[(reduced3_df["Year"] == "2009") | 
#                               (reduced3_df["Year"] == "2010")]
# reduced4_df
```

```python
years_to_analyze=['2009','2010','2011','2012','2013','2014','2015','2016','2017','2018']
Years = []

for index, row in reduced3_df.iterrows():
    for year in years_to_analyze:
        if row["Year"] == year:
            Years.append(row)

Years
```

```python
new_df = pd.DataFrame(Years)
new_df
```

```python
new_df['Month'] = new_df['Month'].replace(['12'], 'Dec')
new_df
```
